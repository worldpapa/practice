맨 처음 왼손 엄지손가락은 * 키패드에 오른손 엄지손가락은 # 키패드 위치에서 시작하며, 엄지손가락을 사용하는 규칙은 다음과 같습니다.

```python
numbers = [1, 3, 4, 5, 8, 2, 1, 4, 5, 9, 5] 
hand = "right"
ans = []
left = [-1]
right = [-2]
for i in numbers:
  if i in [1,4,7]:
    ans.append("L")
    left.append(i)
    print('첫번째', i, "L")
  if i in [3,6,9]:
    ans.append("R")
    right.append(i)
    print('두번째', i, "R")

  if hand == "left" and i in [2,5,8,0]:
    if i == right[-1]-1 or right[-1]+1: 
      ans.append("R")
      print('세번째', i, 'R')
    else:
      ans.append("L")
      print('네번째', i, "L")

  elif hand == "right" and i in [2,5,8,0]:
    if i == (left[-1]-1) or (left[-1]+1): 
      ans.append("L")
      left.append(i)
      
      print('다섯번째', i, "L")
    else:
      ans.append("R")
      right.append(i)
      print('여섯번째', i, 'R')

print(ans)
print(left)
print(right)
```

```python
numbers = [1, 3, 4, 5, 8, 2, 1, 4, 5, 9, 5] 
hand = "right"

ans = ''
#좌표 변경
dic = { 1 : [0,0], 2 : [0,1], 3 : [0,2],
       4 : [1,0], 5 : [1,1], 6 : [1,2],
       7 : [2,0], 8 : [2,1], 9 : [2,2],
       '*' : [3,0], 0 : [3,1], '#' : [3,2]}

    # 시작 위치
left_s = dic['*']
right_s = dic['#']

for i in numbers:
  now = dic[i]
  if i in [1,4,7]:
    ans += 'L'
    left_s = now
    print('첫번째', 'L')
  elif i in [3,6,9]:
    ans += 'R'
    right_s = now
    print('두번째', 'R')
  else: ##2,5,8,0 경우
    left_d = 0
    right_d = 0
    
    for a, b, c in zip(left_s, right_s, now):
      left_d += abs(a-c)
      right_d += abs(b-c)
    
    if left_d > right_d: # 왼손 거리가 더 크다면
      ans += "R"
      right_s = now
      print('세번째', 'R')

    elif left_d < right_d:
      ans += 'L'
      left_s = now
      print('네번째', 'L')
    else: # 두 거리가 같은 경우
      if hand == 'right': #오른손 잡이면
        ans += 'R'
        right_s = now
        print('다섯번째', 'R')
      else:
        ans += 'L'
        left_s = now
        print('여섯번째', 'L')
print(ans)

```
