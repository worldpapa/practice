```python
import itertools

N, K = map(int, input().split())

bagW = []
bagV = []

for i in range(N):
  W, V = map(int, input().split())
  bagW.append(W) #무게를 다 더함
  bagV.append(V)

ap = []

for i in range(len(bagW)):
  nPr = itertools.permutations(bagW, i)
  nPr = list(nPr)
  nPrV = itertools.permutations(bagV, i)
  nPrV = list(nPrV)

  #print("i", i, nPr)
  for j in range(len(nPr)):
    if sum(nPr[j]) <= K:
      #print('j :', j, nPr[j])
      ap.append(sum(nPrV[j]))

ap = set(ap)
ap = list(ap)
print(max(ap))
```

```python
N,K = map(int, input().split())
weight = [0]
gold =[0]
for _ in range(N):
    w, g= map(int, input().split())
    weight.append(w)
    gold.append(g)
    


dp=[[0  for i in range(K+1)] for k in range(N+1)]
for w in range(1, N+1):

    for i in range(1, K+1):
        if i>= weight[w]:
          #  print(weight[w],gold[w],dp[w]  )
                
                dp[w][i]=max(dp[w-1][i], dp[w-1][i-weight[w]]+gold[w])
                # 요게 w-1 인 이유 : 중복 제거
        else:
            
                dp[w][i]=dp[w-1][i]
print(dp[N][K])
```

```python
import sys

N, K = map(int, input().split())
stuff = [[0,0]]
knapsack = [[0 for _ in range(K + 1)] for _ in range(N + 1)] #y축 무게, x축 물건

for _ in range(N):
    stuff.append(list(map(int, input().split())))


#냅색 문제 풀이
for i in range(1, N + 1):
    for j in range(1, K + 1): #j가 1부터 7까지 먼저 훑는다
        weight = stuff[i][0] #i가 1일때 (첫번째 물건일 때)의 무게 6
        value = stuff[i][1] #첫번째 물건일 때의 가치 13
       
        if j < weight: #현재 무게가 6이니까 6보다 작은건 더할 필요가 없으니 위에 값 그대로 (이 물건을 포함안한 전 무게들) 받아옴
            knapsack[i][j] = knapsack[i - 1][j] #weight보다 작으면 위의 값을 그대로 가져온다
        else: #이제 무게 조건 충족하니까 max( (현재 물건 가치 + knapsack[이전 물건][현재 가방 무게 - 현재 물건 무게]) , (knapsack[이전 물건][현재 가방 무게]) )
            knapsack[i][j] = max(value + knapsack[i - 1][j - weight], knapsack[i - 1][j])

print(knapsack[N][K])
```
